```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_sales = pd.read_excel("linechart.xlsx")
df_sales.head()
```

```python
plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"],color="green")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"],color="orange")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"],color="red")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("Product Sales")
plt.xlabel("Finacial Quarter")
plt.ylabel("Revenue")
plt.plot(df_sales["Quarter"],df_sales["Fridge"],color="green",label="Fridge")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"],color="orange",label="Dishwasher")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"],color="red",label="Washing Machine")
plt.legend()                    ## used to add label to graph
plt.show()
```

```python
# create Subplot
plt.figure(figsize=(12,4))

plt.subplot(2,2,1)
plt.plot(df_sales["Quarter"],df_sales["Fridge"],color="green",label="Fridge")
plt.xticks(rotation=45)

plt.subplot(2,2,2)
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"],color="orange",label="Dishwasher")
plt.xticks(rotation=45)

plt.subplot(2,2,3)
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"],color="red",label="Washing Machine")

plt.xticks(rotation=45)
plt.show()
```

```python
total_sale=df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sale
```

```python
total_sale.index
```

```python
plt.pie(total_sale,labels=total_sale.index)
plt.show()
```

```python
plt.pie(total_sale,labels=total_sale.index,autopct="%1.1f%%")
plt.show()
```

```python
plt.pie(total_sale,labels=total_sale.index,autopct="%1.1f%%",explode=(0.1,0,0))     ## explode from origin
plt.show()
```

```python
plt.pie(total_sale,labels=total_sale.index,autopct="%1.1f%%",explode=(0.1,0,0),shadow=True)    ## apply shadow
plt.show()
```

```python
df_sales.plot(kind="bar",x="Quarter")
## rotate the label
plt.xticks(rotation=45)
plt.show()
```
