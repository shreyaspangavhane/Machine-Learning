# Matplotlib
      

 **It is the plooting library of python. It is used to create the 2D, 3D diagram for better visualization                                                                 
 it is the numerical mathematical extension of numpy**

```python
import matplotlib.pyplot as plt           # matplotlib.pyplot is the library
%matplotlib inline                  
# %matplotlib inline -> this is prevent if we want to display any plot or diagram we need to use plt.show()
```

```python
import numpy as np
```

```python
#example
x=np.arange(1,10)
y=np.arange(11,20)
```

# 2D Graph

```python
#ploting of graph

#scatter
plt.scatter(x,y,c="g")                
# x & y is the value  and c is color here g is green color     s is also there for pixel

plt.xlabel("X axis")
plt.ylabel("Y axis")                 #xlabel and ylabel used to give axis name
plt.title("Graph in 2D")       #give the title to graph
```

```python
plt.savefig("Test.png")                        #used to save this fig in device as png file
```

```python
y=x*x
```

```python
#plot function
plt.plot(x,y,"r--",linewidth=3,alpha=0.2)             #  -- line of graph , linewidth is used to increases the width  of line , alpha is used for level of transperancy
```

```python
plt.plot(x,y,"y-o")             #  -o line of graph

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("2D graph")
```

```python
plt.plot(x,y,"bo")             #  doted line of graph
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("2D graph")
```

```python
days=[1,2,3,4,5,6,7]
max_temp=[23,42,35,22,30,30,23]
min_temp=[19,29,19,28,26,20,18]
avg_temp=[21,36,24,21,29,25,20]
plt.plot(x,y)
plt.ylabel("Temp")
plt.xlabel("Days")
plt.title("Weather")

#ploting
plt.plot(days,max_temp,label="Max")
plt.plot(days,min_temp,label="Min")
plt.plot(days,avg_temp,label="Avg")
plt.legend(loc="best",shadow=True)                              #ysed for labeling  and loc is lower left,uper right,upper left,best,
```

**Creating Subplot**

```python
#creating Subplot

plt.subplot(2,2,1)      # subplot(no. of row, no. of col, no. of diagram)    here 1 is 1st diagram
plt.plot(x,y,"g--")

plt.subplot(2,2,2)       #here 2 is the number 2 diagram
plt.plot(x,y,"r*-")

plt.subplot(2,2,3)         #here 3 is the number 3 diagram
plt.plot(x,y,"yo-")
```

```python
plt.savefig("Subplot Graph.png")            #saving 
```

```python
a=np.arange(1,11)                     #plot diagram using arange numpy function
b=6*a+2
plt.plot(a,b)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()                    # make visibillity of the col and row of graph
```

```python
# from x and y co-ordinate making sine and cosin wave graph

x=np.arange(0, 4 * np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Sine graph")
```

```python
#plot graph in one cell using subplot function
plt.subplot(3,3,1)                      #make 3 by 3 part of graph using subplot

x=np.arange(0, 4 * np.pi,0.1)

y_sin=np.sin(x)                   #calculate sin using numpy inbuild function 
plt.plot(x,y_sin)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Sine graph")

plt.subplot(3,3,2)
y_cos=np.cos(x)                  #calculate cos using numpy inbuild function 
plt.plot(x,y_cos)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("cos graph")
```

# Bar

```python
#Bar Plot

x=[5,2,9]
y=[12,4,10]

x1=[3,5,2]
y1=[5,7,11]

plt.bar(x,y)                #having default color as blue 
plt.bar(x1,y1,color="g")              #having Green color\

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Bar graph")
```

```python
# xticks used to change the label of x axis
company=['Google','Amazon','FaceBook','Meta']
revenue=[90,132,76,94]
ypos=np.arange(len(company))
plt.xticks(ypos,company)
plt.bar(ypos,revenue)

plt.ylabel("Revenue")
plt.xlabel("Company")
plt.title("Share")
```

```python
# xticks used to change the label of x axis
company=['Google','Amazon','FaceBook','Meta']
revenue=[90,132,76,94]
profit=[30,45,23,55]
ypos=np.arange(len(company))
plt.xticks(ypos,company)
plt.bar(ypos-0.2,revenue,width=0.4,label="Revenue")           #here ypos-0.2 used to change the position and width used to make 0.4 width that bar
plt.bar(ypos+0.2,profit,width=0.4,label="Profit")
plt.legend()

plt.ylabel("Revenue")
plt.xlabel("Company")
plt.title("Share")
```

```python
plt.savefig("Bar Graph.png")                 #saving graph
```

# Piechart

**.pie(sizes,explode,labels,colors) function has 4 value :-                                                                                 
sizes -> tell us about the how much portion want to cover.                                  
labels -> tell us about the name of that portion                                                        
explode -> tell us about the which portion want to explode means want to out from diagram                        
colors -> tell us about color**


```python
#pie function value of data
labels='AIML','DS','SDE','CC'
sizes=[56,45,60,23]
colors=['lightblue','gold','lightyellow','lightgreen']
explode=(0.1,0,0,0) 

#plot the pie
plt.pie(sizes,explode=explode,colors=colors,labels=labels,autopct="%1.1f%%",shadow=True)

plt.show()         #it does not print the explanation or content only show chart or fig
```

```python
plt.pie(sizes,explode=explode,colors=colors,labels=labels,autopct="%1.1f%%",shadow=True,startangle=45)
#start angle used to change the location of staerting 
plt.show()
```

```python
plt.savefig("Pie Chart.png",bbox_inches="tight")
```
