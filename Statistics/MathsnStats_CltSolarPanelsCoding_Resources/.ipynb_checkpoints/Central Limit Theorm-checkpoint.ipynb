```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("bill_survey.csv")
df.head()
```

```python
df.City.nunique()
```

```python
# print the data of each city
for city, df1 in df.groupby("City"):
    print(city)
    print(df1)
    break
```

```python
# print the mean of each city

sample_mean=[]
for city, df1 in df.groupby("City"):
    sample_mean.append(df1.Bill.mean())

print("Length is:",len(sample_mean))
sample_mean
```

```python
sns.histplot(sample_mean,kde=True)
```
