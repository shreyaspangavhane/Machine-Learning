```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("heights.csv")
df.head()
```

```python
mean=round(df.height.mean(),2)
std=round(df.height.std(),2)
mean,std
```

```python
sns.histplot(df.height,kde=True)
```

```python
#  plot the mean line in histogram
sns.histplot(df.height,kde=True)
plt.axvline(mean,color="red",linestyle="dashed")
```

```python
#  plot the mean and std differnce and addition line in histogram
sns.histplot(df.height,kde=True)
plt.axvline(mean,color="b",linestyle="dashed")

plt.axvline(mean+std,color="r",linestyle="dashed")
plt.axvline(mean-std,color="r",linestyle="dashed")
```

```python
#  plot the mean and 2 std,3 std  differnce and addition line in histogram
sns.histplot(df.height,kde=True)
plt.axvline(mean,color="b",linestyle="dashed")

# mean and std
plt.axvline(mean+std,color="r",linestyle="dashed")
plt.axvline(mean-std,color="r",linestyle="dashed")

# mean with 2* std
plt.axvline(mean+2*std,color="g",linestyle="dashed")
plt.axvline(mean-2*std,color="g",linestyle="dashed")

# mean with 3* std
plt.axvline(mean+3*std,color="black",linestyle="dashed")
plt.axvline(mean-3*std,color="black",linestyle="dashed")
```

```python
left1=round(mean-std,2)
right1=round(mean+std,2)
left1,right1
```

```python
left2=round(mean-2*std,2)
right2=round(mean+2*std,2)
left2,right2
```

```python
left3=round(mean-3*std,2)
right3=round(mean+3*std,2)
left3,right3
```

```python
# calculate the data inside the two region  =>> data between mean and std regiion (means data inside the red lines)
df[(df.height>left1)&(df.height<right1)].head()
```

```python
df[(df.height>left1)&(df.height<right1)].shape
```

```python
# mean with 2 std 
df[(df.height>left2)&(df.height<right2)].head()
```
