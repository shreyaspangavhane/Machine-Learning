```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sample_size=50;
sample=df.sample(sample_size,random_state=100)
sample.head()
```

```python
# calculate the Mean and Std dev of total population
mean_population=df.miles.mean()
std_dev_population=df.miles.std()
```

```python
mean_population,std_dev_population
```

```python
# calculate the Mean and Std dev of sample population
mean_sample=sample.miles.mean()
std_sample=sample.miles.std()
```

```python
mean_sample,std_sample
```

```python
# cal z score for 95 confodence level
import scipy.stats as st

z_score_95_CI=st.norm.ppf(0.975)
z_score_95_CI
```

```python
margin_of_error = z_score_95_CI*(std_dev_population/np.sqrt(sample_size))
margin_of_error
```

```python
lower=mean_sample-margin_of_error
upper=mean_sample+margin_of_error
lower,upper
```
