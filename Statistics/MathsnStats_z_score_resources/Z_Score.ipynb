```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("heights_with_outliers.csv")
df.head(5)
```

```python
mean = df.height.mean()
std = df.height.std()

mean, std
```

```python
sns.histplot(df.height, kde=True)

plt.axvline(mean, color='k', linestyle='dashed', linewidth=1)

# one stadard deviation
plt.axvline(mean + std, color='r', linestyle='dashed', linewidth=1)
plt.axvline(mean - std, color='r', linestyle='dashed', linewidth=1)

# two standard deviation
plt.axvline(mean + 2*std, color='g', linestyle='dashed', linewidth=1)
plt.axvline(mean - 2*std, color='g', linestyle='dashed', linewidth=1)

# three standard deviation
plt.axvline(mean + 3*std, color='b', linestyle='dashed', linewidth=1)
plt.axvline(mean - 3*std, color='b', linestyle='dashed', linewidth=1)
```

```python
#  Calculate the Z score
df['z_score']=(df['height']-mean)/std
df.head(5)
```

```python
# Check outlier
df[(df.z_score<-3)|(df.z_score>3)]
```

```python
## Remove outlier
df_no_outlier=df[(df.z_score>-3)&(df.z_score<3)]
sns.histplot(df_no_outlier.height,kde=True)
```

```python
## Standard Normal Distribution of df with no outlier
sns.histplot(df_no_outlier.z_score)
```
