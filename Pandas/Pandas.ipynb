```python
df = pd.DataFrame({"name": ['Shreyas', 'Rahul', 'Rohit'],
                   "dept": [np.nan, 'Comp', 'Comp'],
                   "City": [pd.NaT, np.nan,'Shirdi']})
```

```python
df
```

```python
df.dropna()          #used to remove Null
```

```python
df   
```

```python
df.dropna(how="all")                   #if all coloumn value is equal to null then it will be delete that coloumn  
```

```python
df.drop_duplicates(subset=['dept'])                        #check in dept column
```

```python
df.drop_duplicates(subset=['dept'],keep="first")            #remove Last term
```

```python
df.drop_duplicates(subset=['dept'],keep="last")      #remove first term
```

```python
df.drop_duplicates(subset=['dept'],keep=False)                  #remove ALL term
```

```python
df.notnull()
```

```python
df.isnull()
```
