```python
import pandas as pd
```

```python
df=pd.read_csv("weather_data.csv",parse_dates=['day'])       # parse_dates => used to make time or date datatype 
df
```

```python
type(df.day[0])
```

```python
df.columns
```

```python
df.set_index('day',inplace=True)
df
```

```python
# Handling missing value by replace NaN to 0
# df.fillna(0,inplace=True)
# df
```

```python
#handle missing value using replace NaN to Average/Mean value
df.fillna({
    "temperature":df.temperature.mean(),
    "windspeed":df.windspeed.mean(),
    "event":"No Event"
})
```

```python
df
```

```python
#handle missing value using by replace missing value to the forward/before value
df.fillna(method="ffill")
```

```python
#  to set the limit for that
df.fillna(method="ffill",limit=1)
```

```python
#handle missing value using by replace missing value to the backword/after value
df.fillna(method="bfill")
```

```python
#handle missing value using by replace missing value to the coloums copy
df.fillna(method="bfill",axis="columns")
```

```python
#handle missing value using by replace missing value to the interpolate/set middle value
df.interpolate()
```

```python
# Dropna is used to drop the data row which consist of missing value
df.dropna()
```

```python
## consist of 3 parameter  any,all,thresh
df.dropna(how="any")            ## delete that row which any of item in row contain NaN
```

```python
df.dropna(how="all")            ## delete that row which has all row contain NaN
```

```python
df.dropna(thresh=2)      # delete that row which has more or equal to 2 NaN
```
