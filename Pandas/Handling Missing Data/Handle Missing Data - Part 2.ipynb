```python
import pandas as pd
```

```python
df=pd.read_csv("weather_data_part_2.csv")
df
```

```python
## Here -99999 is the missing value
```

```python
# replace -99999 with 0
df.replace(-99999,0)
```

```python
# replace -99999 and -88888 with NaN
import numpy as np
df.replace([-99999,-88888],np.nan)
```

```python
# replace all null or missing value(no event also)
df.replace({
    "temperature":-99999,
    "windspeed":[-99999,-88888],
    "event":"no event"
},np.nan)
```

```python
df.replace({
    -99999:np.nan,
    -88888:np.nan,
    "no event":"Sunny"
})                        ## here key is old value and value is new value which you asssigned
```
